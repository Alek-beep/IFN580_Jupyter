```python
def factorial(n):
    if n < 0:
        # Factorial is not defined for negative numbers
        raise ValueError("Factorial is not defined for negative numbers")
    elif n == 0:
        return 1
    else:
        result = 1
        # Loop from 1 up to and including n
        for i in range(1, n + 1):
            result *= i
        return result
```

```python
factorial(5)
```
